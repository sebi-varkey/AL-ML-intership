```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("/content/preprocessing_practice_dataset.csv")
df
```

```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
df[['Age_mean_imputed','Age_median_imputed']]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Experience_Years',y='Monthly_Salary',palette='pastel')
plt.title('Experience vs Salary')
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Age',y='Monthly_Salary',palette='pastel')
plt.title('Age vs Salary')
plt.title('Experience vs Performance Rating')
```
